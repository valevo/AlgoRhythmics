```python
def clean_score(score):
    '''
    - Transposes to key of C
    - Sets simultaneous notes to a chord object (chordifies parts, any problems?)
    - Fixes first bar to be of correct length and fill with rest (eg in Bach chorales)
    - Set time signature denominator to 4?
    - etc
    '''
    key = score.analyze('key')
    diff = m21.interval.Interval(key.tonic, m21.pitch.Pitch('C'))
    
    new_score = m21.stream.Score()
    
    for part in score.parts:
        # check if first bar adds up...
        m0 = part.getElementsByClass(m21.stream.Measure)[0]
        if m0.duration != m0.barDuration:
            # need to pad out the left side with a rest...
            shift = m0.paddingLeft
            r = m21.note.Rest(quarterLength=shift)
            m0.insertAndShift(0, r)
            for element in m0.elements:
                if element not in m0.notesAndRests:
                    element.offset = 0.0
            
            for m in part.getElementsByClass(m21.stream.Measure)[1:]:
                m.offset += shift
        
        new_score.insert(part.offset, part.chordify())
    
    return new_score.transpose(diff)
```

```python
def parse_note_data(part):
    '''
    Returns data in form (beat_stength, pitch_class, octave)
    where pitch_class is -1 for rests, and has (+) operator if part of chord
    '''
    data = []
    d = []
    last_octave = 4
    
    for n in part.flat.notesAndRests:
        bs = n.beatStrength
        if isinstance(n, m21.note.Note):
            oc = last_octave if not n.pitch.octave else n.pitch.octave
            data.append((bs, str(n.pitch.pitchClass), oc))
            last_octave = oc
            print(n.offset, n, data[-1])
            
        elif isinstance(n, m21.note.Rest):
            data.append((bs, '-1', last_octave))
            print(n.offset, n, data[-1])
            
        elif isinstance(n, m21.chord.Chord):
            ps = n.pitches
            for p in ps[:-1]:
                oc = last_octave if not p.octave else p.octave
                data.append((bs, str(p.pitchClass) + '+', oc))
                print(n.offset, n, data[-1])
            
            oc = last_octave if not ps[-1].octave else ps[-1].octave
            data.append((bs, str(ps[-1].pitchClass), oc))
            print(n.offset, n, data[-1])
            
        else:
            print('Something else encountered: ', n)
        
            
    return data
```

```python
offsets = []
for n in test_score.flat.notesAndRests:
    offsets.append(n.offset)
offsets.append(test_score.quarterLength)
print(offsets)
```

```python
for i in range(int(test_score.quarterLength)):        
    elem = list(filter(lambda x: i <= x < (i+1), offsets))
    word = [x%1 for x in elem]
    if len(word) == 0:
        word.insert(0, "_")
        print(word)
        continue
        
    if word[0] != 0.0 or len(word) == 0:
        word.insert(0, "_")
    print(word)
```

```python
def rhythm_word(score):
    offsets = []
    for n in score.flat.notesAndRests:
        offsets.append(float(n.offset))
    offsets.append(score.quarterLength)
    
    for i in range(int(score.quarterLength)):        
        elem = list(filter(lambda x: i <= x < (i+1), offsets))
        word = ['{:.4f}'.format(x%1) for x in elem]
        if len(word) == 0:
            word = ["_"]
            print(word)
            continue

        if word[0] != 0.0 or len(word) == 0:
            word.insert(0, "_")
        print(word)
```

```python
for m in s1.parts[0].getElementsByClass('Measure'):
    m.stripTies().show()
    rhythm_word(m.stripTies())
    
```
