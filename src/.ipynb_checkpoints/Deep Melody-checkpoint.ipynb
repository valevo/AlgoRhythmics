```python
def seqToOneHot(seq, vocabulary, length=None):
    '''
    converts an integer sequence to a one-hot encoding for parsing data
    '''
    if not length:
        length = len(seq)
    
    one_hot = np.zeros((length, vocabulary))
    for i, x in enumerate(seq):
        one_hot[i, x] = 1
    return one_hot
```

```python
''' Load and parse data '''

allBach = m21.corpus.search('bach')
print('Loaded corpus...')
rhythm_data = []   # just holds the rhythmic data
melody_data = []   # holds the melodic data

rhythm_words = {'EOM': 0}                 # dictionary of rhythm tokens
melody_words = {'EOM': 128, 'SOM': 129}   # dictionary of special melody tokens (otherwise midi notes)
r_counter = 1

# iterate over all (Bach) pieces:
for piece in tqdm_notebook(allBach):
    score = piece.parse()
    
    # transcribe to middle C
    dt = 60 - score.analyze('key').tonic.midi
    
    # iterate over all parts:
    for part in score.transpose(dt).parts:
        curr_ts = '4/4'
        # iterate over all measures:
        for m in part.getElementsByClass('Measure'):
            ts = m.getElementsByClass('TimeSignature')
            if ts:
                curr_ts = ts[0].ratioString
                
            if curr_ts not in rhythm_words:
                rhythm_words[curr_ts] = r_counter
                r_counter += 1
                
            r_bar = [rhythm_words[curr_ts]]
            m_bar = [melody_words['SOM']]
            
            for x in m.flat.notes:
                
                # NOTE VALUES
                if isinstance(x, m21.chord.Chord):
                    note = x.root().midi
                else:
                    note = x.pitch.midi
                
                # RHYTHM
                d = x.duration.quarterLength
                if d not in rhythm_words:
                    rhythm_words[d] = r_counter
                    r_counter += 1
                
                r_bar.append(rhythm_words[d])
                m_bar.append(note)
                
            
            r_bar.append(rhythm_words['EOM'])
            m_bar.append(melody_words['EOM'])
            rhythm_data.append(r_bar)
            melody_data.append(m_bar)
                
        
print('Parsing Finished!')
    
r_vocabulary = len(rhythm_words)
m_vocabulary = len(melody_words)

rhythm_ids = dict(zip(rhythm_words.values(), rhythm_words.keys()))
melody_ids = dict(zip(melody_words.values(), melody_words.keys()))

print('Rhythm vocabulary length: {}\nMelody vocabulary length: {}'.format(r_vocabulary, m_vocabulary))
print('Data length: {}'.format(len(rhythm_data)))
```

```python
# building the model!
# RHYTHM
r_latent_dim = 32
```

```python
r_encoder_inputs = Input(shape=(None, r_vocabulary), name='RE_inputs')
r_encoder = LSTM(r_latent_dim, return_state=True, name='RE_lstm')
r_encoder_outputs, r_state_h, r_state_c = r_encoder(r_encoder_inputs)
r_encoder_states = [r_state_h, r_state_c]

r_decoder_inputs = Input(shape=(None, r_vocabulary), name='RD_inputs')
r_decoder_lstm = LSTM(r_latent_dim, return_sequences=True, return_state=True, name='RD_lstm')
r_decoder_outputs, _, _ = r_decoder_lstm(r_decoder_inputs, initial_state=r_encoder_states)
r_decoder_dense = Dense(r_vocabulary, activation='softmax', name='RD_dense')
r_decoder_outputs = r_decoder_dense(r_decoder_outputs)

r_model = Model([r_encoder_inputs, r_decoder_inputs], r_decoder_outputs)
```

```python
r_encoder_model = Model(r_encoder_inputs, r_encoder_states)

r_decoder_states_input_h = Input(shape=(r_latent_dim,))
r_decoder_states_input_c = Input(shape=(r_latent_dim,))
r_decoder_states_inputs = [r_decoder_states_input_h, r_decoder_states_input_c]
r_decoder_outputs, r_state_h, r_state_c = r_decoder_lstm(r_decoder_inputs, r_decoder_states_inputs)
r_decoder_states = [r_state_h, r_state_c]
r_decoder_outputs = r_decoder_dense(r_decoder_outputs)
r_decoder_model = Model([r_decoder_inputs] + r_decoder_states_inputs, [r_decoder_outputs] + r_decoder_states)
```

```python
def encode_rhythm_sequence(seq):
    '''
    Encode the one-hot representation of a musical sequence to latent space.
    '''
    if len(seq.shape) == 2:
        seq = seq.reshape((1, *seq.shape))
    
    return r_encoder_model.predict(seq)

def decode_rhythm_sequence(code, start='4/4', return_states=False):
    '''
    Decode the latent space into a full measure.
    '''
    states_value = code
    
    target_seq = np.zeros((1, 1, r_vocabulary))
    target_seq[0, 0, rhythm_words[start]] = 1
    
    stop = False
    decoded_bar = [rhythm_words[start]]
    
    while not stop:
        output_tokens, h, c = r_decoder_model.predict([target_seq] + states_value)
        
        sampled_token_idx = np.argmax(output_tokens[0, -1, :])
        decoded_bar.append(sampled_token_idx)
        
        if (sampled_token_idx == rhythm_words['EOM'] or len(decoded_bar) > max_input_length):
            stop = True
            
        target_seq = np.zeros((1, 1, r_vocabulary))
        target_seq[0, 0, sampled_token_idx] = 1
        
        states_value = [h, c]
    
    if return_states:
        return decoded_bar, states_value
    else:
        return decoded_bar
```

```python
def parse_seq(sequence, stream=None, pad=True):
    if not stream:
        stream = m21.stream.Stream(key='C')
        
    measure = m21.stream.Measure()
    measure.timeSignature = m21.meter.TimeSignature('4/4')
    
    for word in sequence:
        if word == 1:
            break
        elif word == 0:
            continue
        val = id_to_word[word]
        d = m21.duration.Duration(val[1])
        measure.append(m21.note.Note(val[0], type=d.type))
        
    if pad:
        if measure.duration.quarterLength < 4:
            d = m21.duration.Duration(4 - measure.duration.quarterLength)
            measure.append(m21.note.Rest(type=d.type))
        
    stream.append(measure)
    
    return stream
```

```python
def test_network(seq_idx, text=False, midi=False):
    '''
    Prints a comparison for validation purpose. Options to show text and midi
    '''
    
    input_seq = r_encoder_input_data[seq_idx]
#     print(input_seq)
    encoded_bar = encode_rhythm_sequence(input_seq)
    decoded_bar = decode_rhythm_sequence(encoded_bar, start=rhythm_ids[rhythm_data[seq_idx][0]])
    
#     if text or midi:
#         s_in = parse_seq(input_data[seq_idx - validation_length])
#         s_out = parse_seq(decoded_bar)
    
    print('-- {} --'.format(seq_idx))
    in_s = [rhythm_ids[x] for x in rhythm_data[seq_idx]]
    print('Input sequence:   {}'.format(in_s))
#     if text:
#         s_in.show('text')
#     if midi:
#         s_in.show('midi')
    
    out_s = [rhythm_ids[x] for x in decoded_bar]
    print('Decoded sequence: {}'.format(out_s))
#     if text:
#         s_out.show('text')
#     if midi:
#         s_out.show('midi')
    
```
